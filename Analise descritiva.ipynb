```python
import pandas as pd
ano = '2020'
```

```python
df = pd.read_parquet('s3://pamdas-sandbox/desafio/sinasc/stage/%s/part-00000-2fe0a81b-bac7-4a72-9f70-85a448b6030d-c000.snappy.parquet'%ano)
df2 = pd.read_parquet('s3://pamdas-sandbox/desafio/sinasc/stage/%s/part-00001-2fe0a81b-bac7-4a72-9f70-85a448b6030d-c000.snappy.parquet'%ano)
df = pd.concat([df, df2])
df_pop = pd.read_csv('s3://pamdas-sandbox/desafio/populacao_estados/pop_uf.csv')
```

```python
df_malcon = df[df['IDANOMAL']=='1']
```

```python
df_malcon = df_malcon[['nasc_SIGLA_UF', 'IDANOMAL']].groupby(by='nasc_SIGLA_UF').count()
```

```python
df_malcon = df_malcon.merge(df_pop, left_on='nasc_SIGLA_UF', right_on='uf')
df_malcon
```

```python
df_malcon['proporcao_casos_pop'] = df_malcon['IDANOMAL'] / df_malcon['pop']
df_malcon['casos_10k_hab'] = df_malcon['proporcao_casos_pop'] * 10000
```

```python
df_malcon
```

-------------------------------------

```python
nasc_df = df[['nasc_SIGLA_UF', 'IDANOMAL']].groupby(by='nasc_SIGLA_UF').count()
```

```python
nasc_df['UF'] = nasc_df.index
```

```python
df_malcon = df_malcon.merge(nasc_df, left_on='uf', right_on='UF')
df_malcon
```

```python
df_malcon['casos_nasc'] = df_malcon['IDANOMAL_x'] / df_malcon['IDANOMAL_y']
df_malcon['porc_casos_nasc'] = df_malcon['casos_nasc'] * 100
df_malcon['total_nasc'] = df_malcon['IDANOMAL_y']

df_malcon = df_malcon[['uf', 'estado', 'proporcao_casos_pop','casos_10k_hab', 'casos_nasc', 'porc_casos_nasc', 'total_nasc']]
```

---------------------------

```python
tipo_cid = pd.read_csv('s3://pamdas-sandbox/desafio/cid/grupo_cid.csv')
tipo_cid
```

```python
df_malform = df[df['IDANOMAL']=='1']
```

```python
import re
for nome_grupo, codigos in tipo_cid.values:
    df_malform[nome_grupo] = df_malform['CODANOMAL'].map(lambda codes: bool(re.search(codigos, str(codes))))
```

```python
df_malform
```

```python
df_tipos = df_malform.groupby(by=['nasc_SIGLA_UF']).sum()
```

```python
df_tipos
```

```python
df_malcon = df_malcon.merge(df_tipos, left_on='uf', right_on='nasc_SIGLA_UF')
```

```python
for i in df_malcon.index:
    for nome_grupo in ["Espinha bífida","Outras malformações congênitas do sistema nervoso","Malformações congênitas do aparelho circulatório","Fenda labial e fenda palatina","Ausência atresia e estenose do intestino delgado","Outras malformações congênitas do aparelho digestivo","Testiculo não-descido","Outras malformações do aparelho geniturinário","Deformidades congênitas do quadril","Deformidades congênitas dos pés","Outras malformações e deformidades congênitas do aparelho osteomuscular","Outras malformações congênitas","Anomalias cromossômicas não classificadas em outra parte"]:
        df_malcon.loc[i, nome_grupo+'prop'] = 100*df_malcon.loc[i, nome_grupo]/df_malcon.loc[i, 'total_nasc']
```

```python
df_malcon.head()
```

```python
df_malcon_result = df_malcon[["uf","estado","casos_nasc","Espinha bífidaprop","Outras malformações congênitas do sistema nervosoprop","Malformações congênitas do aparelho circulatórioprop","Fenda labial e fenda palatinaprop","Ausência atresia e estenose do intestino delgadoprop","Outras malformações congênitas do aparelho digestivoprop","Testiculo não-descidoprop","Outras malformações do aparelho geniturinárioprop","Deformidades congênitas do quadrilprop","Deformidades congênitas dos pésprop","Outras malformações e deformidades congênitas do aparelho osteomuscularprop","Outras malformações congênitasprop","Anomalias cromossômicas não classificadas em outra parteprop"]]
```

```python
df_malcon_result.rename(columns={"uf": "UF",
                                 "estado": "Estado",
                                 "casos_nasc": "Casos_entre_nascidos"}).to_csv("s3://pamdas-sandbox/desafio/sinasc/tabela_malformacao_%s.csv"%ano, index=False)
```
